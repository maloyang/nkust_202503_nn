- 有預設值的Function


````python
def function4(m=1, n=10):
    sum = m+n
    return sum
````

````python
function4()
````

````python
function4(m=5)
````

````python
function4(50)
````

````python
def function5(m, n=10):
    sum = m+n
    return sum
````

````python
function5(5, 8)
````

````python
def func5(a, b=1, c=2, d=3):
    return a+b+c+d
````

````python
func5(5) #a=5, b=1, c=2, d=3
````

````python
func5(5, d=10)
````

````python
5+1+2+10
````
